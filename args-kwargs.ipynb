### 1. Write a function called average_score that calculates the average of an arbitrary number of scores

```python
# *scores allows us to handle any number of arguments
def average_score(*scores):
    if len(scores) == 0:
        return 0
    average_score = sum(scores) / len(scores)
    return average_score
```

```python
average_score(20, 30, 40)
```

```python
average_score(100, 50)
```

```python
average_score()
```

### 2. Write a function describe_pet that accepts a pet name and type as required args, but can also take any number of keyword details

```python
# **details becomes a dictionary of all the keyword arguments
def describe_pet(name, animal_type, **details):
    print(f"Pet name: {name} of type {animal_type}")

    if details:
        print('Details: ')
        for key, value in details.items():
            print(f" {key}: {value}")
```

```python
describe_pet('Milo', 'shitzu')
```

```python
describe_pet('Milo', 'shitzu', age=3)
```

```python
describe_pet('Milo', 'shitzu', age=3, color='wheatish')
```

### 3. Write a function make_sandwich that accepts any number of ingredients and any number of optional preferences

```python
# when having both * and **, the order matters, tuple should come first before the dictionary
def make_sandwich(*ingredients, **preferences):
    for item in ingredients:
        print(f" {item}")
    
    if preferences:
        print("Optional preferences:")
        for k, v in preferences.items():
            print(f"{k}: {v}")
        print(preferences)
```

```python
make_sandwich("bread", "cheese", "patty", dressing1='mayo', dressing2='honey-mustard')
```

```python
make_sandwich("bread", "cheese", "patty")
```
