Creating the decorator function `performance_metrics`

```python
def performance_metrics(base_fn):
    def enhanced_fn(*args, **kwargs):
        start_time = time.time()
        print(f"Calling function: {base_fn}")
        result = base_fn(*args, **kwargs)
        end_time = time.time()
        print(f"Time taken to process: {(end_time - start_time):.2f}")
        return result
    return enhanced_fn
```

```python
@performance_metrics
def get_users_dec(num_users):
    print("before sleep..")
    time.sleep(2)
    print(f"getting all the {num_users} users..")
    return num_users
```

```python
get_users_dec(10)
```
